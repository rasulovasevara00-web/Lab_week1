# Problem 3

```
scores = [45, 67, 82, 59]
def pass_fail(score):
  if score >= 60 :
    return "Pass"
  else:
    return "Fail"
for score in scores:
  result = pass_fail(score)
  print(f"Input: {score} Output: {result}")
```

# Problem 4

```
sizes = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]
for i in range(len(sizes)):
  print(sizes[i],prices[i])
increasing = True
for i in range(1,len(prices)):
  if prices[i] <= prices[i-1]:
    increasing = False
if increasing:
  print("Positive trend.")
else:
  print("Negtaive trend.")
```

# Problem 5

```
study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]
increases = []
for i in range(1, len(scores)):
    increases.append(scores[i] - scores[i - 1])
average_increase = sum(increases) / len(increases)
predicted_score = scores[-1] + average_increase

print("Average increase per 2 hours:", average_increase)
print("Predicted score for 10 hours of study:", predicted_score)
```

# Problem 6


```
scores = [35, 58, 62, 74, 89, 45]

for score in scores:
    if score < 50:
        category = "Low"
    elif score <= 70:
        category = "Medium"
    else:
        category = "High"

    print(score, "-", category)
```
